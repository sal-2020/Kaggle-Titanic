```python
# Time to model !!!
```

```python
X_train = df[(df['IsTrainSet']==1)].drop(columns=['PassengerId', 'Survived', 'IsTrainSet'])
X_test = df[(df['IsTrainSet']==0)].drop(columns=['PassengerId', 'Survived', 'IsTrainSet'])
y_train = df.loc[df['IsTrainSet']==1,['Survived']]
y_train = y_train['Survived']
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
X_train = StandardScaler().fit(X_train).transform(X_train)
X_test = StandardScaler().fit(X_test).transform(X_test)

LogR = LogisticRegression()
LogR.fit(X_train, y_train)
LogR
```

```python
print('Logistic Regresion Results')
print(confusion_matrix(y_train, LogR.predict(X_train)))
print(classification_report(y_train, LogR.predict(X_train)))
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import GridSearchCV

svc = SVC()
svc.fit(X_train, y_train)
svc
```

```python
print('Support Vector Machines Results 1')
print( confusion_matrix(y_train, svc.predict(X_train) ) )
print( classification_report(y_train, svc.predict(X_train) ) )
```

```python
svc_hyperparameters  = {'C':[.01,.1,1,10,100], 'gamma':[.00001,.0001,.001,.01,.1,1]}
grid_svc = GridSearchCV(SVC(), svc_hyperparameters)
grid_svc.fit(X_train, y_train)
grid_svc.best_estimator_
```

```python
print('Support Vector Machines Results - Grid Search')
print( confusion_matrix(y_train, grid_svc.predict(X_train) ) )
print( classification_report(y_train, grid_svc.predict(X_train) ) )
```

```python
from sklearn.neighbors import KNeighborsClassifier

KNN = KNeighborsClassifier(n_neighbors=3)
KNN.fit(X_train, y_train)
KNN
```

```python
print('K Nearest Neighbors Results')
print( confusion_matrix(y_train, KNN.predict(X_train) ) )
print( classification_report(y_train, KNN.predict(X_train) ) )
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree

dt = DecisionTreeClassifier(max_depth=6)
dt.fit(X_train, y_train)
dt
```

```python
print('Decision Tree Results')
print( confusion_matrix(y_train, dt.predict(X_train) ) )
print( classification_report(y_train, dt.predict(X_train) ) )
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=150, max_depth=6)
rf.fit(X_train, y_train)
rf
```

```python
print('Random Forest Results')
print( confusion_matrix(y_train, rf.predict(X_train) ) )
print( classification_report(y_train, rf.predict(X_train) ) )
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb = GradientBoostingClassifier(n_estimators=250)
gb.fit(X_train, y_train)
gb
```

```python
print('Gradient Boosting Results')
print( confusion_matrix(y_train, gb.predict(X_train) ) )
print( classification_report(y_train, gb.predict(X_train) ) )
```

```python
# Use the Gradient Boosting model

gb_pred = gb.predict(X_test)

results1 = df.loc[df['IsTrainSet']==0,['PassengerId']]
results1['Survived'] = gb_pred
results1['Survived'] = results1['Survived'].astype(int)
results1
```

```python
results1.to_csv('Titanic_Submission_2.csv', index=False)
```
